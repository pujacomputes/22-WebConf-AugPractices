```python
import torch
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
import glob
```

```python
def make_plots(stats,model_id):
    
    fig, axes = plt.subplots(4, 2, figsize=(15, 12))
    axes[0][0].plot(stats['epoch_loss'], color="tab:blue", label="Loss")
    axes[0][0].set_ylabel("Running Loss")
    ax_1 = axes[0][0].twinx()
    ax_1.plot(stats["knn_acc"], color="tab:orange", label="Acc",alpha = 0.5)
    ax_1.set_ylabel("KNN Acc")
    axes[0][0].legend(loc='upper left')
    ax_1.legend(loc='upper right')

    axes[1][0].plot(stats['backbone_sim'], color="tab:cyan", label='Backbone_Sim')
    axes[1][0].legend()
    axes[1][0].set_ylabel("Backbone Sim")

    axes[2][0].plot(stats['encoder_sim'], color="tab:green", label='Projector_Sim')
    axes[2][0].legend()
    axes[2][0].set_ylabel("Sim")

    axes[3][0].plot(stats['predictor_sim'], color="tab:purple", label='Predictor_Sim')
    axes[3][0].legend()
    axes[3][0].set_ylabel("Sim")

    axes[3][0].legend()
    axes[3][0].set_ylabel("Sim")
    axes[3][0].set_xlabel("Batches/Epochs")

    axes[0][1].plot(stats['epoch_loss'], color="tab:blue", label="Loss")
    axes[0][1].set_ylabel("Running Loss")
    ax_1 = axes[0][1].twinx()
    ax_1.plot(stats["knn_acc"], color="tab:orange", label="Acc",alpha = 0.5)
    ax_1.set_ylabel("KNN Acc")
    axes[0][1].legend(loc='upper left')
    ax_1.legend(loc='upper right')

    axes[1][1].plot(stats['backbone_norm'], color="tab:cyan", label='Backbone_Norm')
    axes[1][1].legend()
    axes[1][1].set_ylabel("Backbone Sim")
    ax_1 = axes[1][1].twinx()
    ax_1.plot(stats['backbone_std'],label='Backbone_Std')

    axes[2][1].plot(stats['encoder_norm'], color="tab:green", label='Projector_Norm')
    axes[2][1].legend()
    axes[2][1].set_ylabel("Sim")
    ax_1 = axes[2][1].twinx()
    ax_1.plot(stats['encoder_std'],label='Projector_Std')

    axes[3][1].plot(stats['predictor_norm'], color="tab:purple", label='Predictor_Norm')
    axes[3][1].legend()
    axes[3][1].set_ylabel("Sim")
    ax_1 = axes[3][1].twinx()
    ax_1.plot(stats['predictor_std'],label='Predictor_Std')
    axes[3][1].set_xlabel("Batches/Epochs")

    fig.suptitle(model_id);
    
    return fig
```
